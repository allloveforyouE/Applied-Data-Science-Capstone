```python
import pandas as pd
import numpy as np

# =====================================================================
# PHẦN 1: TẢI THẲNG DỮ LIỆU CHUẨN ĐÃ QUA API TỪ MÁY CHỦ CLOUD CỦA IBM
# =====================================================================
print("Đang tải dữ liệu chuẩn hóa từ IBM Cloud...")

# Đường dẫn file dữ liệu tĩnh phục vụ riêng cho bài Lab khi API sập
data_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_1.csv"

# Đọc trực tiếp vào biến data_falcon9 để khớp với cấu trúc bài Lab của bạn
data_falcon9 = pd.read_csv(data_url)

print("Kích thước dữ liệu gốc:", data_falcon9.shape)
print("Số giá trị rỗng trong cột PayloadMass TRƯỚC khi xử lý:", data_falcon9['PayloadMass'].isnull().sum())

# =====================================================================
# TASK 3: DEALING WITH MISSING VALUES (TÍNH MEAN VÀ THAY THẾ)
# =====================================================================

# Bước 1: Tính giá trị trung bình (mean) của cột PayloadMass
payload_mass_mean = data_falcon9['PayloadMass'].mean()
print(f"\n-> Giá trị trung bình (mean) của PayloadMass: {payload_mass_mean:.2f} kg")

# Bước 2: Sử dụng hàm replace() để thay thế các giá trị khuyết thiếu (np.nan) bằng giá trị mean
data_falcon9['PayloadMass'] = data_falcon9['PayloadMass'].replace(np.nan, payload_mass_mean)

# Kiểm tra lại kết quả
missing_count = data_falcon9['PayloadMass'].isnull().sum()
print("-> Số giá trị rỗng trong cột PayloadMass SAU khi xử lý:", missing_count)

# =====================================================================
# PHẦN 3: XUẤT FILE SANG DẠNG CSV THEO YÊU CẦU
# =====================================================================
if missing_count == 0:
    data_falcon9.to_csv('dataset_part_1.csv', index=False)
    print("\n--- ĐÃ XUẤT FILE 'dataset_part_1.csv' THÀNH CÔNG")
else:
    print("\nVẫn còn giá trị rỗng, vui lòng kiểm tra lại cột dữ liệu.")
```
